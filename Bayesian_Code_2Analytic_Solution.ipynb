```python
#Solution
def solution(K, Res):
    k = range(1, 100)
    global d_0
    d_0 = []
    #NDots = linspace(0, D, Res)
    #ndots = len(NDots)*len(NDots)
    for i in range(len(d4)): #escala los puntos del espacio
        T = []
        for j in range (K):# j escala los vecinos
            teo = d4[i][j]**2 / k[j] 
            T.append(teo)
        T2 = (sqrt(sum(T))) 
        d_0.append(1/(T2*T2*pi)) #This is divided in order to get n_0
    #print T3
solution(12, 200)
 
print len(d_0)
print d_0[0:10]
#print amax(T2)
#print sum(T)
```
