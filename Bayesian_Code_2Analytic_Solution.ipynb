```python
#Sigma Estimator
def sigma_estimator(K, Res):
    k = range(1, 100)
    global sigma_2
    sigma_2 = []
    #NDots = linspace(0, D, Res)
    #ndots = len(NDots)*len(NDots)
    for i in range(len(d4)): #escala los puntos del espacio
        S = []
        for j in range (K):# j escala los vecinos 
            sigma1 = (-3*d4[i][j]**2 + k[j]*d_0[i]**2)  # Ojo los signos están alrevez
            S.append(sigma1)
        S2 = ((sum(S))) 
        sigma = sqrt(-K*d_0[i]**4)/(2*S2)
        sigma_2.append(pi*sigma**2) 
    #print type(sigma), dtype(S[1]), type(S), dtype(S2), type(sigma_2)
    
sigma_estimator(8, 200)
 
print len(sigma_2), dtype(sigma_2[1])
#print amax(T2)
#print sum(T)
sigmaT = []
for i in range(len(sigma_2)):
    sigmaT.append(sigma_2[i].real)
    if sigmaT[i] == nan:
        sigmaT[i] = 1000
print len(sigmaT), dtype(sigmaT[2])
print amin(sigmaT), amax(sigmaT)
```
