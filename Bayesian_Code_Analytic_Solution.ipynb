```python
import numpy as pn 
from scipy import *
import matplotlib as plt
import math 
from random import *
```

```python
def Init_Distribution(nx, ny):
    global X
    global Y
    X = np.random.random(1000)*nx
    Y = np.random.random(1000)*ny
    figure(num=None, figsize=(9.5,9))
    xlabel('$\mathrm{x}$', fontsize = 35)
    ylabel('$\mathrm{y}$', fontsize = 35)
    title('$\mathrm{Random}$ $\mathrm{Cartesian}$ $\mathrm{Distribution}$ ', fontsize = 40)
    tick_params(axis='both', which='major', labelsize=18)
    scatter(X, Y)
    ylim([0, ny])
    xlim([0, nx])
    savefig('RandomCartesianDistribution.png')

Init_Distribution(1000, 1000)


       
```

```python
def Init_Distribution_Polar(nR):
    global R
    global theta
    R = np.random.random(1000)*nR
    theta = (np.random.random(1000)+1)*randrange(0, 360)
    figure(num=None, figsize=(9.5,9))
    xlabel('$\mathrm{x}$', fontsize = 35)
    ylabel('$\mathrm{y}$', fontsize = 35)
    title('$\mathrm{Random}$ $\mathrm{Polar}$ $\mathrm{Distribution}$', fontsize = 40)
    tick_params(axis='both', which='major', labelsize=18)
    scatter(R*cos(theta), R*sin(theta))
    xlim([-120, 120])
    ylim([-120, 120])
    savefig('RandomPolarDistribution.png')
       
Init_Distribution_Polar(100)
#print R*cos(theta), R*sin(theta)
```

```python
# k = Number of Neighbors, D = Dimension of the space we assume here X = Y | ++POLAR COORDINATES++ 
def NeighborsPolar(k, D):
    global d4Polar
    d4Polar = []
    for i in range(-D, D):# Be carefull with this -D this aplies to the Polar case or where the inital Distribution have negative points
        for j in range(-D, D):
            d = sqrt((R*cos(theta)-i)**2 + (R*sin(theta)-j)**2)# check this equation 
            d2 = sorted(d)
            d3 = d2[0:k]
            d4Polar.append(d3)
    #return d4Polar
    
NeighborsPolar(8, 100)
```

```python
# k = Number of Neighbors, D = Dimension of the space we assume here X = Y | CARTESIAN COORDINATES
def Neighbors(k, D):
    global d4
    d4 = []
    for i in range(0, D):
        for j in range(D):
            d = sqrt((X-i)**2 + (Y-j)**2)
            d2 = sorted(d)
            d3 = d2[0:k]
            d4.append(d3)
    #return d4
    
Neighbors(8, 1000)

#d4 imprime así: [ [k vecinos]i j , [k vecinos]i, j + 1 .....]

```

```python
# Analytical Solution,put this into a function
k = range(1, 100)
T3 = []
for i in range (0, 1000000): #escala los puntos del espacio
    global T2
    T = []
    for j in range (0, 8):# j escala los vecinos
        teo = d4[i][j]**2 / k[j] # + (d4[i][1]**2 / k[1])) 
        T.append(teo)
    T2 = 100. - (sqrt(sum(T))) # this is divided for give the right weight in the histogram
    T3.append(T2*T2*pi)
 
print len(T3)
print T3[0:10]
print amax(T3)
```
