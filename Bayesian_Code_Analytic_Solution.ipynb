```python
# k = Number of Neighbors, D = Dimension of the space we assume here X = Y CARTESIAN coordinates, Make this for polar coordinates 
def NeighborsPolar(k, D):
    global d4Polar
    d4Polar = []
    for i in range(1):
        for j in range(2):
            d = sqrt((R*cos(theta)-i)**2 + (R*sin(theta)-j)**2)# check this equation 
            d2 = sorted(d)
            d3 = d2[0:k]
            d4Polar.append(d3)
    return d4Polar
    
NeighborsPolar(2, 10)
```

```python
# Analytical Solution
k = range(1, 100)
f = open('density.txt', 'w')
T3 = []
for i in range (0, 100): #escala los puntos del espacio
    global T2
    T = []
    for j in range (0, 8):# j escala los vecinos
        teo = d4[i][j]**2 / k[j] # + (d4[i][1]**2 / k[1])) 
        T.append(teo)
    T2 = sqrt(sum(T))
    T3.append(T2)
    f.write(str(T2) + "  " + "\n")
    
f.close()
len(T3)
```
